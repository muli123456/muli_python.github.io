```python
import numpy as np
from matplotlib import pyplot as plt
import math
from matplotlib.ticker import MultipleLocator
```

```python
def emittance(x):
    return 3.74*(10**8)/((x**5)*(np.exp((1.44*(10**4)/((x*255)-1)))))
```

```python
x = np.linspace(0,40)
fig,ax = plt.subplots()
ax.plot(x, emittance(x),'r',label= 'Earth T = 255k')
plt.xlabel('Wavelength $(um)$')
plt.ylabel('Eλ* $(W\,m^{-2}\,\mu m^{-1}$)')
plt.ylim(0,15)
plt.xlim(0, np.max(x))
ax.xaxis.set_major_locator( MultipleLocator(10))
ax.xaxis.set_minor_locator( MultipleLocator(2))
ax.yaxis.set_major_locator( MultipleLocator(5))
ax.yaxis.set_minor_locator( MultipleLocator(1))
ax.set_title('Emittance in Earth')
```

```python
Radiance_lambda = emittance(x)/ math.pi
```

```python
x = np.linspace(0,40)
fig,ax = plt.subplots()
ax.plot(x, Radiance_lambda,'r',label= 'Earth T = 255k')
plt.xlabel('Wavelength $(um)$')
plt.ylabel('Lλ* $(W\,m^{-2}\,\mu m^{-1}$)')
plt.xlim(0, np.max(x))
plt.ylim(0, 5)
ax.xaxis.set_major_locator( MultipleLocator(10))
ax.xaxis.set_minor_locator( MultipleLocator(2))
ax.yaxis.set_major_locator( MultipleLocator(1))
ax.yaxis.set_minor_locator( MultipleLocator(0.2))
ax.set_title('Blackbody Raidance in Earth')
```
